```python
def generate_n_doable_examples(n, model, tokenizer):
    examples = []
    iterator = tqdm(range(n))
    while len(examples) < n:
        ex = sample_example(tokenizer)
        for k, v in ex.base.items():
            if v is not None and isinstance(v, torch.Tensor):
                ex.base[k] = v.to(model.device)
        for k, v in ex.src.items():
            if v is not None and isinstance(v, torch.Tensor):
                ex.src[k] = v.to(model.device)
        logits_base = model(**ex.base).logits[0, -1]
        logits_src = model(**ex.src).logits[0, -1]
        if (
            logits_base[ex.base_label] > logits_base[ex.src_label]
            and logits_src[ex.src_label] > logits_src[ex.base_label]
        ):
            examples.append(ex)
            iterator.update(1)
    return examples
```

```python
# make dataset
total_steps = 100
trainset = generate_n_doable_examples(total_steps, gpt, tokenizer)
evalset = generate_n_doable_examples(50, gpt, tokenizer)
```

```python
def intervention_config(intervention_site, layer, num_dims=1):
    config = pv.IntervenableConfig([
        {
            "layer": layer,
            "component": intervention_site, # "mlp" or "attn"
            "intervention_type": pv.LowRankRotatedSpaceIntervention, #low-rank, rotated subspace intervention
            "low_rank_dimension": num_dims,
        }
    ])
    return config
```

```python
# loss function
loss_fct = torch.nn.CrossEntropyLoss()

def calculate_loss(logits, label):
    """Calculate cross entropy between logits and a single target label (can be batched)"""
    shift_labels = label.to(logits.device)
    loss = loss_fct(logits, shift_labels)
    return loss
```

```python
# intervention settings
stats = []
num_layers = gpt.config.num_hidden_layers
example_stimuli = sample_example(tokenizer)
len_stimuli = example_stimuli.base['input_ids'].shape[1]

# loop over layers and positions
for layer in range(num_layers):
    for position in range(len_stimuli):
        print(f"layer: {layer}, position: {position}")

        # set up intervenable model
        config = intervention_config("block_output", layer, 1)
        intervenable = pv.IntervenableModel(config, gpt)
        intervenable.set_device(device)
        intervenable.disable_model_gradients()

        # set up optimizer
        optimizer_params = []
        for k, v in intervenable.interventions.items():
            try:
                optimizer_params.append({"params": v.rotate_layer.parameters()})
            except:
                pass
        optimizer = torch.optim.Adam(optimizer_params, lr=1e-3)
        scheduler = get_linear_schedule_with_warmup(
            optimizer,
            num_warmup_steps=int(0.1 * total_steps),
            num_training_steps=total_steps,
        )

        # training loop
        iterator = tqdm(trainset)
        for example in iterator:

            # forward pass
            _, counterfactual_outputs = intervenable(
                example.base,
                [example.src],
                {"sources->base": position},
            )

            # loss
            logits = counterfactual_outputs.logits[:, -1]
            loss = calculate_loss(logits, torch.tensor([example.src_label]).to(device))
            iterator.set_postfix({"loss": f"{loss.item():.3f}"})

            # backward
            loss.backward()
            optimizer.step()
            scheduler.step()

        # eval
        with torch.no_grad():
            iia = 0
            data = []
            iterator = tqdm(evalset)
            for example in iterator:
                # forward
                original_outputs, counterfactual_outputs = intervenable(
                    example.base,
                    [example.src],
                    {"sources->base": position},
                    output_original_output=True
                )

                # calculate iia
                logits = counterfactual_outputs.logits[0, -1]
                base_logits = original_outputs.logits[0, -1]
                probs = logits.log_softmax(dim=-1)
                base_probs = base_logits.log_softmax(dim=-1)
                
                data.append({
                    "src_label": example.src_label,
                    "base_label": example.base_label,
                    "loss": loss.item(),
                    "p_base": probs[example.base_label].item(),
                    "p_src": probs[example.src_label].item(),
                    "base_p_base": base_probs[example.base_label].item(),
                    "base_p_src": base_probs[example.src_label].item(),
                    "layer": layer,
                    "pos": position
                })
                # odds is "odds_ratio": sum([d['base_p_base'] - d['base_p_src'] + d['p_src'] - d['p_base'] for d in data]) / len(data),
            
            summary = {
                "iia": sum([d['p_src'] > d['p_base'] for d in data]) / len(data), # proportion of times intervention led to preference for src label
                "iia-flip": sum([d['p_src'] > d['p_base'] for d in data if d['base_p_base'] > d['base_p_src']]) / len(data), # proportion of times intervention led to preference for src label when base model preferred base label
                "odds_ratio": sum([d['base_p_base'] - d['base_p_src'] + d['p_src'] - d['p_base'] for d in data]) / len(data), # magnitude of intervention effect
                "eval_loss": sum([d['loss'] for d in data]) / len(data), # mean cross-entropy for this layer/position
                "layer": layer, 
                "position": position
            }

            # stats
            stats.append(summary)
            print(f"odds: {summary['odds_ratio']}")
df = pd.DataFrame(stats)
```

```python
df["layer"] = df["layer"].astype(int)
df["pos"] = df["position"].astype(int)
df["IIA"] = df["iia"].astype(float)

custom_labels = ["EOS", "The", "THING", "was", "good/bad", "but", "not"]
# breaks = list(range(len_stimuli))
breaks = [0, 1, 3, 4, 5, 6,7]
print(breaks)

pos_to_label = dict(zip(breaks, custom_labels))

df["pos_label"] = df["pos"].map(pos_to_label)
df["pos_label"] = pd.Categorical(
    df["pos_label"],
    categories=custom_labels[::-1],  # reverse here if you want EOS at top
    ordered=True
)

plot = (
    ggplot(df, aes(x="layer", y="pos_label"))
    + geom_tile(aes(fill="odds_ratio"))
    + scale_fill_cmap("Purples")
    + xlab("layers")
    + ylab("")
    + theme(figure_size=(5, 3))
    + theme(axis_text_y=element_text(hjust=1))
    + ggtitle("Trained Intervention (DAS)")
)

plot.show()

```
